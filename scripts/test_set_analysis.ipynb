```python
chunk_names_path = '/home/asr/datasets/indicvoices_versioned_test_manifests/v1/chunk_names.txt'
test_root = '/home/asr/speech-datasets/indicvoices/eval-data'
```

```python
with open(chunk_names_path) as reader:
    names = reader.read().strip().splitlines()
```

```python
import glob
```

```python
all_errs = glob.glob(f'{test_root}/**/*.err',recursive=True)
chunk_names_err_all = [f.split('/')[-1].split('.')[0] for f in all_errs]
```

```python
chunk_names_test = [f.split('.')[0] for f in names]
```

```python
final_list = []
for f in set(chunk_names_test) & set(chunk_names_err_all):
    for g in all_errs:
        if f in g:
            final_list.append(g)
```

```python
print('\n'.join(final_list))
```

```python
len(final_list)
```

```python
for f in final_list:
    with open(f) as reader:
        lines = reader.read()
    # print(f.split('/')[6])
    with open(f"err/{f.split('/')[6]}.txt",'a') as writer:
        print('\n',file=writer)
        print(f.split('/')[-1],file=writer)
        print(lines,file=writer)
    # with open()
```
